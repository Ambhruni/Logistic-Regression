1.What is Simple Linear Regression (SLR)? Explain its purpose?

->Simple Linear Regression (SLR) is a statistical method used to study the relationship between two variables â€” one independent (predictor) variable and one dependent (response) variable.

It helps us predict the value of the dependent variable based on the value of the independent variable.

The equation of the regression line is:

Y=a+bX+e

Where:

Y = Dependent variable

X = Independent variable

a = Intercept

b = Slope

e = Error term

Purpose of Simple Linear Regression:

Prediction:
To predict the value of one variable based on another.

Understanding Relationships:
To identify whether and how strongly two variables are related.

Trend Estimation:
To model and understand trends over time or other continuous variables.



2.What are the key assumptions of Simple Linear Regression?

->The key assumptions of Simple Linear Regression (SLR) ensure that the model is valid and the results are reliable. These assumptions help us check whether the relationship between the variables can be accurately represented by a straight line.

1. Linearity

The relationship between the independent variable (X) and dependent variable (Y) should be linear.

This means a change in X leads to a proportional change in Y.

2. Independence of Errors

The residuals (differences between actual and predicted values) should be independent of each other.

one observationâ€™s error should not influence anotherâ€™s.

Homoscedasticity (Constant Variance of Errors)

The variance of residuals should be constant across all levels of X.

If the spread of residuals increases or decreases with X, the assumption is violated

4. Normality of Errors

The residuals (errors) should be normally distributed.

This is important for making valid confidence intervals and hypothesis tests.

5. No (or minimal) Measurement Error in X

The independent variable (X) should be measured accurately.

Large measurement errors in X can lead to biased estimates of the slope.

6. No Significant Outliers

Outliers can distort the regression line and make it unrepresentative.

3.Write the mathematical equation for a simple linear regression model and explain each term?

->Mathematical Equation for Simple Linear Regression

Y=a+bX+e

where;

Y:Dependent Variable

The variable we want to predict or explain

X:Independent Variable

The variable used to predict the value of Y

a:Intercept

The expected value of Y when X = 0. It shows where the regression line crosses the Y-axis.

b:Slope

Shows how much Y changes for a one-unit increase in X.
- If b > 0, Y increases with X (positive relationship).
- If b < 0, Y decreases with X (negative relationship).

e:Error Term

The difference between the actual value of Y and the predicted value from the regression line. It accounts for random variation not explained by X.

9.Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.


```python
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 5, 4, 5])
model = LinearRegression()
model.fit(X, Y)
slope = model.coef_[0]
intercept = model.intercept_
print("Slope (b):", slope)
print("Intercept (a):", intercept)
Y_pred = model.predict(X)
print("Predicted values:", Y_pred)
```
